Este notebook tem como objetivo realizar a coleta dos dados para o trabalho.

O setor escolhido foi o de mineração e de Siderurgia.

Link da B3 que define os setores das empresas listadas: https://www.b3.com.br/pt_br/produtos-e-servicos/negociacao/renda-variavel/empresas-listadas.htm

As ações que compoem o setor de mineração:
- COMPANHIA BRASILEIRA DE ALUMÍNIO (CBAV) - Extração de minério de alumínio. serviços de usinagem. caldeira e montagem de maquinário.
- CSN MINERAÇÃO S.A. (CMIN3) - Mineração e comercialização de minério de ferro. infraestrutura de transporte portuária; geração de energia e participação em sociedades com objeto social compatível com o da Companhia.
- VALE S.A. (VALE3) - Mineração

Açoes do setor de Siderurgia: 

- CIA SIDERURGICA NACIONAL (CSNA3) - A Companhia Siderúrgica Nacional atua nos setores de mineração. siderurgia. logística. cimento e energia.
- GERDAU S.A. (GGBR4) - Participação e Administração
- USINAS SID DE MINAS GERAIS S.A. (USIM5) - Venda de produtos siderúrgicos. como Chapas Grossas. Laminados a quente. a frio. placas. galvanizados. entre outros.


Correlação Setorial:
Mineração e siderurgia estão intimamente ligadas, pois a matéria-prima produzida pelo setor de mineração, como o minério de ferro, é fundamental para a produção no setor de siderurgia. Isso cria uma correlação natural entre os desempenhos das empresas desses setores.


```python
# Libaries
import pandas
import yfinance as yf
```

```python
# Criando lista com os ticks das ações
tickers = ["AURA33.SA", "CBAV3.SA", "CMIN3.SA", "VALE3.SA", "CSNA3.SA", "GGBR4.SA", "USIM5.SA"]


dataframes = {}
for tick in tickers:
    data = yf.Ticker(tick)
    df = data.history(period="max")
    dataframes[tick] = df


dataframes
```

```python
# Salvando dataframes em arquivos csv
for key, value in dataframes.items():
    value.to_csv(f"data\\df_{key.split('.')[0]}.csv")
```
